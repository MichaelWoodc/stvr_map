```python
import os

IGNORE_DIRS = {".git", "venv", "env", ".venv", "__pycache__"}
IGNORE_EXTENSIONS = {".gif"}

def list_all_paths(root="."):
    dirs = []
    files = []

    for current_path, dirnames, filenames in os.walk(root):
        # Filter directories in-place so os.walk never enters them
        dirnames[:] = [
            d for d in dirnames
            if d not in IGNORE_DIRS and "env" not in d.lower()
        ]

        # Collect directories
        for d in dirnames:
            dirs.append(os.path.join(current_path, d))

        # Collect files (skip GIFs)
        for f in filenames:
            ext = os.path.splitext(f)[1].lower()
            if ext in IGNORE_EXTENSIONS:
                continue
            files.append(os.path.join(current_path, f))

    return {
        "directories": dirs,
        "files": files
    }

if __name__ == "__main__":
    result = list_all_paths(".")
    print("\n--- DIRECTORIES ---")
    for d in result["directories"]:
        print(d)

    print("\n--- FILES ---")
    for f in result["files"]:
        print(f)
```

```python
import os
import pandas as pd
import geopandas as gpd

ROOT = "./data_files/parcels"

IGNORE_DIRS = {".git", "venv", "env", ".venv", "__pycache__"}
IGNORE_EXTENSIONS = {".gif"}

def explore_file(path):
    ext = os.path.splitext(path)[1].lower()

    print("\n==============================")
    print(f"Exploring: {path}")
    print("==============================")

    try:
        # --- CSV ---
        if ext == ".csv":
            df = pd.read_csv(path, low_memory=False)
            print("Type: CSV")
            print("Columns:")
            for c in df.columns:
                print(" -", c)
            print("\nSample rows:")
            print(df.head(5))

        # --- GEOJSON ---
        elif ext == ".geojson":
            gdf = gpd.read_file(path)
            print("Type: GeoJSON")
            print("Columns:")
            for c in gdf.columns:
                print(" -", c)
            print("CRS:", gdf.crs)
            print("Geometry type:", gdf.geometry.geom_type.unique())
            print("\nSample rows:")
            print(gdf.head(5))

        # --- SHP ---
        elif ext == ".shp":
            gdf = gpd.read_file(path)
            print("Type: Shapefile")
            print("Columns:")
            for c in gdf.columns:
                print(" -", c)
            print("CRS:", gdf.crs)
            print("Geometry type:", gdf.geometry.geom_type.unique())
            print("\nSample rows:")
            print(gdf.head(5))

        # --- KML ---
        elif ext == ".kml":
            gdf = gpd.read_file(path)
            print("Type: KML")
            print("Columns:")
            for c in gdf.columns:
                print(" -", c)
            print("CRS:", gdf.crs)
            print("Geometry type:", gdf.geometry.geom_type.unique())
            print("\nSample rows:")
            print(gdf.head(5))

        # --- DBF ---
        elif ext == ".dbf":
            # geopandas can read DBF directly
            gdf = gpd.read_file(path)
            print("Type: DBF")
            print("Columns:")
            for c in gdf.columns:
                print(" -", c)
            print("\nSample rows:")
            print(gdf.head(5))

        # --- PRJ / SHX / CPG ---
        elif ext in {".prj", ".shx", ".cpg"}:
            print(f"Type: {ext.upper()} (support file for shapefile)")

        # --- ZIP ---
        elif ext == ".zip":
            print("Type: ZIP archive (likely shapefile or district data)")

        else:
            print("Unknown or unsupported file type.")

    except Exception as e:
        print("Error reading file:", e)


def explore_all(root):
    for current_path, dirnames, filenames in os.walk(root):
        # ignore unwanted dirs
        dirnames[:] = [
            d for d in dirnames
            if d not in IGNORE_DIRS and "env" not in d.lower()
        ]

        for f in filenames:
            ext = os.path.splitext(f)[1].lower()
            if ext in IGNORE_EXTENSIONS:
                continue

            full_path = os.path.join(current_path, f)
            explore_file(full_path)


if __name__ == "__main__":
    explore_all(ROOT)
```
